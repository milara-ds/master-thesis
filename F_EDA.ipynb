```python
df_h1= df.iloc[:,38:400]
df_h1 = df_h1.dropna(how='all', axis=1)
df_h1.info()
```

## **ANOVA test**

```python
df = pd.read_csv("C:/Users/Milara/...") 
df.info()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax = sns.boxplot(x="Area", y="Average Score", data=df , ax = ax)
```

```python
from scipy.stats import f_oneway

eng = df[df['Area'] == 'Engineering']['For_Prom']
bus = df[df['Area'] == 'Business']['For_Prom']
cre = df[df['Area'] == 'Creative Studies']['For_Prom']
hea = df[df['Area'] == 'Health']['For_Prom']
soc = df[df['Area'] == 'Social Sciences']['For_Prom']
bul = df[df['Area'] == 'Built Environment']['For_Prom']

f_oneway(eng, bus, cre, hea, soc, bul)
```
